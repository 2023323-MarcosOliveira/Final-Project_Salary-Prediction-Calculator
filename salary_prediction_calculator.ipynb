# 2.1 Importing required libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## 2.2 Loading the dataset

```python
df = pd.read_csv("survey_results_public.csv")
df.head()
```

## 2.3 Checking the dataset info

```python
df.info()
```

## 2.4 Selecting and renaming
Selecting these columns and renaming `ConvertedCompYearly` to Salary

```python
df = df[["Country", "EdLevel", "YearsCodePro", "Employment", "ConvertedCompYearly"]]
df = df.rename({"ConvertedCompYearly": "Salary"}, axis=1)
df.head()
```

## 2.5 Displaying all Salaries

Displaying all salaries where it is `not null`

```python
df = df[df["Salary"].notnull()]
df.head()
```

## 2.6 Reading the New Dataset

```python
df.info()
```

## 2.7 Removing Missing Values

Removing rows from the DataFrame that contain `missing (NaN) values`

```python
df = df.dropna()
```

Confirming that there are no more `null values`in the DataFrame

```python
df.isnull().sum()
```

## 2.8 Filtering the DataFrame

Filtering the DataFrame for respondents who are `employed full time`

```python
df = df[df["Employment"] == "Employed, full-time"]
df = df.drop("Employment", axis=1)
```

```python
df.info()
```
